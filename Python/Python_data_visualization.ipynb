# Python Data Visualization libraries


## A. Matplotlib 
> https://matplotlib.org/

- Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. 
- Matplotlib makes easy things easy and hard things possible.
- `matplotlib.pyplot` is a collection of functions that make matplotlib work like MATLAB.


## B. Seaborn
> https://seaborn.pydata.org/

- Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.
- If Matplotlib “tries to make easy things easy and hard things possible”, Seaborn tries to make a well-defined set of hard things easy too.


## C. Bokeh
> https://docs.bokeh.org/en/latest/index.html

- Bokeh is an **interactive** visualization library for modern web browsers.

```python
## step1
source = ColumnDataSource(ColumnDataSource.from_df(data1))
tooltips = [('sample', '@s')]

## step2
p_non_ref = figure(height=500, width=800, 
                   x_axis_label = 'Single-nucleotide variants', 
                   y_axis_label = 'Insertions and Deletions',
                   title = "Number of called variants(hom_var & het)",
                   tooltips = tooltips)

## step3
p_non_ref.scatter(x='n_SNV', y='n_indel',source=source, color = "grey", alpha = 0.6, size = 10)

## step4
p_non_ref.title.text_font_size = '15pt'
p_non_ref.xaxis.formatter = BasicTickFormatter(use_scientific = False)
p_non_ref.yaxis.formatter = BasicTickFormatter(use_scientific = False)

show(p_non_ref)
```

```python
## step1
source = ColumnDataSource(ColumnDataSource.from_df(data1))
tooltips = [('sample', '@s'), ('family','@fam'), ('n_SNV', '$x'), ('n_Indel', '$y')]
families = list(set(data1['fam']))

## step2
p_by_group = figure(height=500, width=800, 
                   x_axis_label = 'Single-nucleotide variants', 
                   y_axis_label = 'Insertions and Deletions',
                   title = "Number of called variants(hom_var & het) - by family",
                   tooltips = tooltips)

## step3
p_by_group.scatter(x='n_SNV', y='n_indel', source=source, alpha = 0.6, size = 10,
                   color=factor_cmap('fam', viridis(9), families), legend = 'fam')

## step4
p_by_group.legend.location = "top_left"
p_by_group.title.text_font_size = '15pt'

p_by_group.xaxis.formatter = BasicTickFormatter(use_scientific = False)
p_by_group.yaxis.formatter = BasicTickFormatter(use_scientific = False)

show(p_by_group)
```

```python
p= column(p_non_ref, p_by_group)
#save(p, 'Figures/bokeh_plot.html', title='Sample QC')
export_png(p, 'Figures/bokeh_plot.png', 500, 1400)
```
